```python
filenames = [i[:-4] for i in os.listdir(dir_params)]

list_df = {}

file_paths = [os.path.join(dir_params,x) for x in os.listdir(dir_params)]

for i in range(len(file_paths)):
    if file_paths[i][-3:] =='csv':
        list_df[filenames[i]] = pd.read_csv(rf'{file_paths[i]}', sep=';' )

```

```python
with pd.ExcelWriter('parameters/merged_file.xlsx') as writer:  
    for df in list(list_df.keys()):
        list_df[df].to_excel(writer, sheet_name=f'{df}')
```
